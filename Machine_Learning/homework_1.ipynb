### Домашнее задание к занятию "Классификация: Логистическая регрессия и SVM"

Имеются данные adult.csv (см. в материалах к занятию).

Целевой переменной является уровень дохода income (крайний правый столбец).

Описание признаков можно найти по ссылке http://www.cs.toronto.edu/~delve/data/adult/adultDetail.html

Вам необходимо построить модели логистической регрессии и SVM, которые предсказывает уровень дохода человека.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('adult.csv', na_values='?')
df.head()
```

```python
print('Количество пропусков')
for col in df.columns:
    pct_missing = df[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
cat_features = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country', 'class']
cont_features = ['age', 'fnlwgt', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week']
```

```python
df[[x for x in df.columns if x in cont_features]]
```

```python
df[[x for x in df.columns if x in cat_features]]
```

```python
df_dummy = pd.get_dummies(data=df[[x for x in df.columns if x in cat_features]])
df_dummy
```

```python
data = df_dummy.merge(df[[x for x in df.columns if x in cont_features]], right_index=True, left_index=True)
data
```

```python
data_class = pd.get_dummies(data=df['income']).iloc[:,0]
data_class
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, data_class, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=42)
lr.fit(X_train, y_train)
```

```python
predictions = pd.DataFrame([lr.predict(X_test), y_test]).T
predictions.columns = ['predict', 'test']
predictions
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, lr.predict(X_test))
```

```python
from sklearn.svm import SVC
clf = SVC()
clf.fit(X_train, y_train) 
```

```python
predictions = pd.DataFrame([clf.predict(X_test), y_test]).T
predictions.columns = ['predict', 'test']
predictions
```

```python
accuracy_score(y_test, clf.predict(X_test))
```
