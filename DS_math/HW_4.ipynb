### Уровень 0:


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.



```python
import numpy as np
from scipy.misc import derivative
import matplotlib.pyplot as plt 
```

```python
def f(x):
    return x**5 + 4*np.sin(2*x) + np.cos(3*x + 3)
```

```python
derivative(f, x0=1.0, dx=1e-6)
```

```python
derivative(f, x0=1.0, dx=1e-6, n=2)
```

### Уровень 1:


Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
def f(x):
    return (np.sin(2*x+1))**5
```

```python
x = np.linspace(-5, 5, num=200)
y1 = [derivative(f, x0=x0, dx=1e-6) for x0 in x]
y2 = [derivative(f, x0=x0, dx=1e-6, n=2) for x0 in x]
```

```python
plt.figure(figsize=(12,6))
plt.plot(x, y1, label='первая производная')
plt.plot(x, y2, label='вторая производная')
plt.title('График первой и второй производной функции f(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
```
