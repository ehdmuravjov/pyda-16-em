**Оценка качества вина по его физико-химическим показателям с помощью машинного обучения**

1.	Введение
Построение модели для определения качества вина по его физико-химическим показателям является не только крайне интересной задачей для специалистов и любителей в области машинного обучения, но также будет полезна для виноделов, которые стремятся улучшить качество своего вина, и поможет им корректировать процесс производства вина для улучшения баланса физико-химических показателей. 
Нашему вниманию представлены данные о красных и белых португальских винах "Винью Верде". К сожалению, из-за проблем с конфиденциальностью и логистикой доступны только физико-химические показатели и оценки качества вина на основе сенсорных данных (например, нет данных о сортах винограда, марке вина, продажной цене вина и т. д.).
Рассмотренные модели были основаны на задаче классификации.

2.	Описание данных
Физико-химические показатели в исходных данных следующие: летучая и нелетучая кислотности, лимонная кислота, остаточный сахар, хлориды, свободный и связанный диоксиды серы, плотность, мера кислотности водных растворов рН, сульфаты, процентное содержание алкоголя и цвет вина. 

3.	Предобработка данных
В процессе предобработки данных выяснилось, что пропущенных значений немного (меньше 0,1 % для некоторых параметров). Минимальный балл качества 3, средний 5.8, высший 9, больше всего данных о винах среднего качества (оценка в 6 баллов). Белых вин в несколько раз больше чем красных вин. В данных есть выбросы по следующим показателям: остаточный сахар и диоксид серы (свободный).

4.	Модели
Было рассмотрено 3 модели машинного обучения: одна модель, основанная на методе к-ближайших соседей (KNN), и две модели на основе случайного леса со стандартными и с подобранными гиперпараметрами.

5.	Планы
В будущем хотелось бы рассмотреть не только задачу классификации, но еще и задачу регрессии на следующих моделях: логистическая регрессия и модель регрессии на основе случайного леса (RandomForestRegressor).
