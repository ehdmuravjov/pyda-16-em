Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import expon
from scipy.stats import t
```

```python
x = np.linspace(-10, 10, 100)
```

```python
fig, ax = plt.subplots(1,1, figsize=(15, 8))
ax.plot(x, norm.pdf(x), label='normal distribution')
ax.plot(x, norm.pdf(x, loc=0, scale=1.5), label='normal distribution loc=0, scale 1.5')
ax.plot(x, norm.pdf(x, loc=5, scale=2), label='normal distribution loc=5, scale 2')
ax.plot(x, norm.pdf(x, loc=-5, scale=5), label='normal distribution loc=-5, scale 5')
ax.plot(x, norm.pdf(x, loc=-3, scale=0.8), label='normal distribution loc=-3, scale 0.8')
ax.legend()
```

```python
fig, ax = plt.subplots(1,1, figsize=(15, 8))
ax.plot(x, expon.pdf(x), label='exponential distribution', linestyle='dashed')
ax.plot(x, expon.pdf(x, loc=0, scale=1.5), label='exponential distribution, loc=0, scale 1.5',  linestyle='dashed')
ax.plot(x, expon.pdf(x, loc=5, scale=2), label='exponential distribution, loc=5, scale 2',  linestyle='dashed')
ax.plot(x, expon.pdf(x, loc=-5, scale=5), label='exponential distribution, loc=-5, scale 5',  linestyle='dashed')
ax.plot(x, expon.pdf(x, loc=-3, scale=0.8), label='exponential distribution, loc=-3, scale 0.8',  linestyle='dashed')
ax.legend()
```

```python
df = 2.74
```

```python
fig, ax = plt.subplots(1,1, figsize=(15, 8))
ax.plot(x, t.pdf(x, df), label='Student distribution')
ax.plot(x, t.pdf(x, df, loc=0, scale=1.5), label='Student distribution loc=0, scale 1.5')
ax.plot(x, t.pdf(x, df, loc=5, scale=2), label='Student distribution loc=5, scale 2')
ax.plot(x, t.pdf(x, df, loc=-5, scale=5), label='Student distribution loc=-5, scale 5')
ax.plot(x, t.pdf(x, df, loc=-3, scale=0.8), label='Student distribution loc=-3, scale 0.8')
ax.legend()
```

```python
import math
```

```python
N = 54 + 4 + 28
p1 = 54 / N
p2 = 4 / N
p3 = 28 / N
```

```python
# количество информации несёт сообщение о цвете светофора вычислим по формуле Шеннона
-(p1 * math.log2(p1) + p2 * math.log2(p2) + p3 * math.log2(p3))
```
